### 레드팀 포지션, 블루팀 포지션 나눠서 결측 대체

```python
red_pos=train[['Player_1_position','Player_2_position','Player_3_position','Player_4_position','Player_5_position']]
blue_pos=train[['Player_6_position','Player_7_position','Player_8_position','Player_9_position','Player_10_position']]
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler, MinMaxScaler
```

```python
# 숫자값으로 인코딩
# 라벨 인코더 돌릴려고 일단 ?처리
for i in red_pos.columns:
    red_pos[i][red_pos[i].isna()]='?'
for i in blue_pos.columns:
    blue_pos[i][blue_pos[i].isna()]='?'
```

```python
le=LabelEncoder()
for i in range(5):
    red_pos.iloc[:,i]=le.fit_transform(red_pos.iloc[:,i])
    blue_pos.iloc[:,i]=le.fit_transform(blue_pos.iloc[:,i])
```

### 팀별로 포지션 겹치지 않으므로 로우의 합이 15가 되어야됨 (1+2+3+4+5)

```python
red_pos[red_pos.isin([0]).any(axis=1)]
blue_pos[blue_pos.isin([0]).any(axis=1)]
```

```python
for i in range(len(red_pos)):
    red_pos.iloc[i,].replace(0,15-red_pos.iloc[i,].sum(),inplace=True)
    blue_pos.iloc[i,].replace(0,15-blue_pos.iloc[i,].sum(),inplace=True)
```

```python
red_pos[red_pos.isin([0]).any(axis=1)]
```

```python
blue_pos[blue_pos.isin([0]).any(axis=1)]
```

```python
train[['Player_1_position','Player_2_position','Player_3_position','Player_4_position','Player_5_position']]=red_pos[['Player_1_position', 'Player_2_position', 'Player_3_position',
       'Player_4_position', 'Player_5_position']]
```

```python
train[['Player_6_position', 'Player_7_position', 'Player_8_position',
       'Player_9_position', 'Player_10_position']]=blue_pos[['Player_6_position', 'Player_7_position', 'Player_8_position',
       'Player_9_position', 'Player_10_position']]
```

### 챔피언+포지션 합친 인덱스 번호 부여

```python
train=train[['teamVictory', 'Player_1_pick', 'Player_2_pick','Player_3_pick','Player_4_pick', 'Player_5_pick',
      'Player_6_pick', 'Player_7_pick', 'Player_8_pick', 'Player_9_pick', 'Player_10_pick']]
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
target=train.iloc[:,0]
data=train.drop('teamVictory',axis=1)
```

```python
# 훈련 데이터와 테스트 데이터 분리
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.2, random_state=42)

# 랜덤 포레스트 모델 생성 및 학습
rf = RandomForestClassifier()
rf.fit(X_train, y_train)

# 모델 예측 및 평가
y_pred = rf.predict(X_test)
accuracy = rf.score(X_test, y_test)
print("Accuracy:", accuracy)
```

```python
# 로지스틱 회귀 모델 생성 및 학습
lr = LogisticRegression()
lr.fit(X_train, y_train)

# 모델 예측 및 평가
y_pred = lr.predict(X_test)
accuracy = lr.score(X_test, y_test)
print("Accuracy:", accuracy)
```
